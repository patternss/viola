```python
# Varmista, että olet asentanut ollama-kirjaston
# pip install ollama

import ollama
import sys

model_name = "phi4-mini"

#ollama.pull(model_name)


def main():
    """
    Pääfunktio yksinkertaiselle Ollama-chatbotille.
    """
    model_name = "phi4-mini"
    
    print(f"Chatbot on käynnissä. Käytetään mallia: {model_name}.")
    print("Kirjoita 'poistu' tai 'exit' lopettaaksesi keskustelun.")

    # Keskusteluhistoria, joka lähetetään mallille jokaisella käänteellä
    messages = []

    try:
        # Testataan ensin yhteys Ollama-palvelimeen
        print("\nTestataan yhteyttä Ollama-palvelimeen...")
        ollama.pull(model_name)
        print(f"Malli '{model_name}' on ladattu tai saatavilla.")
        print("-" * 20)
        
    except ollama.OllamaAPIError as e:
        print(f"Virhe Ollama-palvelimen kanssa: {e}", file=sys.stderr)
        print("Varmista, että Ollama on käynnissä ja että 'phi4-mini' -malli on ladattu (ollama run phi4-mini).", file=sys.stderr)
        sys.exit(1)
    try:
        while True:
            # Pyydä käyttäjän syötettä
            user_input = input("Sinä: ").strip()

            if user_input.lower() in ["/poistu", "/exit"]:
                print("Chatbot sulkeutuu. Kiitos keskustelusta!")
                break

            if not user_input:
                continue

            # Lisää käyttäjän viesti keskusteluhistoriaan
            messages.append({'role': 'user', 'content': user_input})

            # Keskustele Ollama-palvelimen kanssa
            # Stream-asetus on True, jotta vastaus tulostuu suoratoistona
            # sanasta sanaan
            response_stream = ollama.chat(model=model_name, messages=messages, stream=True)

            print(f"{model_name.capitalize()}: ", end="")
            
            full_response = ""
            for chunk in response_stream:
                content = chunk['message']['content']
                print(content, end="", flush=True)
                full_response += content

            # Lisää mallin vastaus keskusteluhistoriaan
            messages.append({'role': 'assistant', 'content': full_response})
            
            print("\n")

    except ollama.OllamaAPIError as e:
        # Käsittele API-virheet, kuten mallin puuttuminen tai palvelimen saavuttamattomuus
        print(f"Virhe Ollama-palvelimen kanssa: {e}", file=sys.stderr)
        print("Varmista, että Ollama on käynnissä ja että 'phi4-mini' -malli on ladattu (ollama run phi4-mini).", file=sys.stderr)
    except Exception as e:
        print(f"Tapahtui odottamaton virhe: {e}", file=sys.stderr)

if __name__ == "__main__":
    main()


```
